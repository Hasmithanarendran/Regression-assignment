```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
independent= dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent= dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train
```

```python
from sklearn.svm import SVR #import svr from svm 
regressor=SVR(kernel="rbf",C=5000) #using radial bias function and creating a model
regressor.fit(X_train, Y_train)
```

```python
regressor.intercept_ 
```

```python
y_pred = regressor.predict(X_test) 
```

```python
from sklearn.metrics import r2_score #cross checking actual and predicted data
r_score=r2_score(Y_test,y_pred)
r_score
```

```python
import pickle
filename="final_svm.sav"
```

```python
pickle.dump(regressor,open(filename,'wb'))
```

```python
loaded_model=pickle.load(open("final_svm.sav",'rb'))
result=loaded_model.predict([[12,27,1,1,0]])
```

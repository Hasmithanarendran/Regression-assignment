```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
independent= dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent= dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor #importing decision tree for this algorithm
regressor=DecisionTreeRegressor(criterion='squared_error', splitter='best')
regressor.fit(X_train,Y_train)
```

```python
import matplotlib.pyplot as plt #importing this to show graph according to decision tree output
from sklearn import tree
tree.plot_tree(regressor) 
plt.show()
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test, y_pred)
r_score
```
